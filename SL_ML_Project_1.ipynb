```python
import pandas as pd

import warnings

warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('train.csv')

df_train
```

```python
df_train.info()
```

```python
df_train.isna().any().sum()
```

```python
df_train.dtypes
```

## Columns with 'object' datatype

```python
#data_type = df_train.dtypes
for i in df_train.columns:
    if df_train[i].dtypes =='object':
        print(i)
```

## Lebel Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for cols in df_train.columns:
    if df_train[cols].dtype == 'object':
        df_train[cols] = le.fit_transform(df_train[cols])
```

```python
df_train.info()
```

## Columns with zero varinace

```python
df_train.var() == 0.0
#data_type = df_train.dtypes
for i in df_train.columns:
    if df_train[i].var() ==0:
        print(i)
```

## Dropping values with zero variance

```python
df_train_le = df_train.drop(df_train.var()[df_train.var() == 0.0].index.values,axis = 1)
```

```python
df_train_le
```

## Test dataset

```python
df_test = pd.read_csv('test.csv')

df_test
```

```python
df_test.isna().any().sum()
```

```python
for cols in df_test.columns:
    if df_test[cols].dtypes=='object':
        df_test[cols] = le.fit_transform(df_test[cols])
```

## Droping zero variance columns 

```python
df_test_le = df_test.drop(['X11','X93','X107','X233','X235','X268','X289','X290','X293','X297','X330','X347'],axis =1)

df_test_le
```

## Data Splitting

```python
X_train = df_train_le.drop(['y','ID'] ,axis =1)
y_train = df_train_le['y']
X_train.shape#y_train.shape
```

```python
X_test = df_test_le.drop(['ID'] ,axis =1)
#y_train = df_train_le['y']
X_test.shape
```

## PCA for dimension reduction

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 0.95)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
#df_train_pca,df_test_pca
```

```python
from sklearn.ensemble 
from xgboost import XGBRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
grid_values = {'n_estimators': [75,80,85,95], 'learning_rate': [0.13,0.135,0.14,0.2,0.18], 'max_depth': [1,2,3,4,5]}
grid_xgb = GridSearchCV(xgb.XGBRegressor(), param_grid = grid_values, cv=50, scoring = 'r2')
grid_xgb.fit(X_train_pca, y_train)
predict_xgb = grid_xgb.predict(X_test_pca)

print('Mean score matrix: ', grid_xgb.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_xgb.best_params_)
print('Grid best score (accuracy): ', grid_xgb.best_score_)
```

### We see that accuracy is less ,so we will try to improve it using scaling the dataset

```python
from sklearn.preprocessing import MinMaxScaler
scalling = MinMaxScaler().fit(X_train)
X_train_scalled = scalling.transform(X_train)
test_scalled = scalling.transform(X_test)
```

## PCA on scalled data

```python
pca = PCA(n_components= 0.95)
X_train_pca_n = pca.fit_transform(X_train_scalled)
X_test_pca_n = pca.transform(test_scalled)
#df_train_pca,df_test_pca
```

## Model Implementation

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
grid_values = {'n_estimators': [75,80,85,95], 'learning_rate': [0.13,0.135,0.14,0.2,0.18], 'max_depth': [1,2,3,4,5]}
grid_xgb = GridSearchCV(xgb.XGBRegressor(), param_grid = grid_values, cv=10, scoring = 'r2')
grid_xgb.fit(X_train_pca_n, y_train)
predict_xgb = grid_xgb.predict(X_test_pca_n)

print('Mean score matrix: ', grid_xgb.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_xgb.best_params_)
print('Grid best score (accuracy): ', grid_xgb.best_score_)
```
